<h3>Scraping the Wikipedia page for the table of postal codes of Canada</h3>
<div>BeautifulSoup Library of Python is used for web scraping of table from the Wikipedia. The title of the webpage is printed to check if the page has been scraped successfully or not. Then the table of postal codes of Canada is printed.<div>

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'html5lib')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab, raw=True)
```

```python
table_contents = []
table = soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text == 'Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = row.span.text.split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

df = pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})

df.head()
```

<h2> Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada</h2>

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

<h2>Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods in Canada</h2>

```python
lat_lon.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
df_completed = pd.merge(df, lat_lon, on='PostalCode')
df_completed.head()
```

<h2>Visualizing all the Neighbourhoods of the above data frame using Folium</h2>

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

<h2>The notebook from here includes the Clustering and the plotting of the neighbourhoods of Canada which contain Toronto in their Borough</h2>

<h3>Getting all the rows from the data frame which contains Toronto in their Borough.</h3>

```python
df_completed = df_completed[df_completed['Borough'].str.contains('Toronto', regex=True)]
print(df_completed.shape)
df_completed.head()
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_completed['Latitude'], df_completed['Longitude'], df_completed['Borough'], df_completed['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

<h2>Using KMeans clustering for the clsutering of the neighbourhoods</h2>

```python
k = 4
toronto_clustering = df_completed.drop(['PostalCode', 'Borough', 'Neighborhood'], 1)
kmeans = KMeans(n_clusters = k, random_state = 0).fit(toronto_clustering)
kmeans.labels_
```

```python
df_completed.insert(0, 'Cluster Labels', kmeans.labels_)
df_completed
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color schema for the cluster 
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map

markers_colors = []
for lat, lon, neighborhood, cluster in zip(df_completed['Latitude'], df_completed['Longitude'], df_completed['Neighborhood'], df_completed['Cluster Labels']):
    label = folium.Popup('Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7
    ).add_to(map_clusters)
    
map_clusters
```
